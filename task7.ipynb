Tasks 7: Boolean Arrays and Masks

1. Filter values greater than the mean. 

2. Count how many elements satisfy a condition. 

3. Replace negative values with zero.

```python
np.random.seed(42)
arr = np.random.randn(20) * 10 + 50  

print("Original Array:\n", np.round(arr, 2))
print("-" * 40)
```

```python
mea = np.mean(arr)
filtered_greater = arr[arr > mea]

print(f"Task 1.\nArray Mean: {mea:.2f}")
print(f"Values > Mean: \n{filtered_greater}")
```

```python
condition_mask = arr > 0
count_positive = np.sum(condition_mask) 

print(f"\nTask 2. \nNumber of positive elements: {count_positive}")
```

```python
arr[arr < 0] = 0

print(f"\nTask 3. \nArray after replacing negatives with zero:\n {arr}")
```

Tasks 7: Boolean Arrays and Masks (enhanced difficulty level)

1. Given a NumPy array of real numbers, compute the mean and standard deviation.

2. Create a boolean mask for values greater than (mean + one standard deviation). 

3. Extract the elements that satisfy this condition and compute their proportion relative to the array size. 

4. Replace all values below the mean with the median of the array (without using explicit loops)



```python
mu = np.mean(arr)
sigma = np.std(arr)
print(f"Task 1 \n", f"{mu:.2f}", f"{sigma:.2f}")
```

```python
threshold = mu + sigma
mask = arr > threshold

print(f"\nTask 2. \nBoolean Mask:\n {mask}")
```

```python
extracted_elements = arr[mask]
proportion = np.mean(mask)

print(f"\nTask 3. \nExtracted Elements:\n {np.round(extracted_elements, 2)}")
print(f"\nProportion relative to array size: {proportion:.2%} ({len(extracted_elements)} out of {arr.size})")
```

```python
arr_median = np.median(arr)
print(f"\nTask 4. \nArray Median: {arr_median:.2f}")

# Using boolean indexing to replace values without explicit loops
arr[arr < mu] = arr_median

print(f"\nModified Array (values < mean replaced by median):\n {np.round(arr, 2)}")
```
