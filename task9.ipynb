Coding Tasks 9: Sorting

1. Sort a random array.

2. Sort rows based on the second column.

3. Find top-5 largest values using argsort.

Coding Tasks 9: Sorting (enhanced difficulty level)

1. Generate a 2D NumPy array of shape (8 x 5) with random values.

2. Sort each row independently in descending order.

3. Sort the rows based on the values in the second column.

4. Identify the index of the row with the highest total sum.

```python
import numpy as np
```

```python
arr = np.random.randint(0, 100, 10)
sorted_arr = np.sort(arr)

print(f"Task 1 \nOriginal:", arr)
print("Sorted:", sorted_arr)
```

```python
arr = np.random.randint(0, 50, (5, 3))

sorted_rows = arr[arr[:, 1].argsort()]

print("Task 2\nOriginal array:\n", arr)
print("Sorted by second column:\n", sorted_rows)
```

```python
arr = np.random.randint(0, 100, 20)

top5_indices = np.argsort(arr)[-5:]
top5_values = arr[top5_indices]

print("Task 3\nArray:", arr)
print("Top 5 values:", top5_values)
```

```python
arr = np.random.randint(0, 100, (8, 5))
print("Task 2.1\nOriginal array:\n", arr)
```

```python
sorted_rows = np.sort(arr, axis=1)[:, ::-1]

print("Task 2.2\nRows sorted descending:\n", sorted_rows)
```

```python
sorted_by_second = arr[arr[:, 1].argsort()]

print("Task 2.3\nRows sorted by second column:\n", sorted_by_second)
```

```python
row_sums = arr.sum(axis=1)
max_row_index = np.argmax(row_sums)

print("Task 2.4\nRow sums:", row_sums)
print("Index of row with highest sum:", max_row_index)
```
