1. Use fancy indexing to reorder rows

```python
print("Task 1.1:")
matrix = np.arange(1, 13).reshape(4, 3)
print("Original Matrix:\n", matrix)

# We want to reorder the rows from [0, 1, 2, 3] to [3, 0, 2, 1]
row_indices = [3, 0, 2, 1]
reordered_matrix = matrix[row_indices]
print("\nReordered Matrix:\n", reordered_matrix)
```

2. Extract random samples from an array.

```python
print("Task 1.2: Random Sample:")
data_array = np.array(['A', 'B', 'C', 'D', 'E', 'F', 'G'])

random_indices = np.random.randint(0, len(data_array), size=3)
random_samples = data_array[random_indices]
print(f"Random Indices: {random_indices}")
print(f"Extracted Samples: {random_samples}")
```

1. Create a 1D NumPy array of length 30 with random integers between O and 100.

```python
print("Task 2.1:")
arr = np.random.randint(0, 101, size=30)
print("Original Array:\n", arr)
```

2. Using fancy indexing, select elements located at prime-number indices.

```python
prime_numbers = []
def is_prime(n):
    limit = int(n**0.5) + 1
    
    for i in range(2, limit):
        if n % i == 0:
            return False
    return True

for i in range(2, 31):
    if is_prime(i)==True:
        prime_numbers.append(i)
    else:
        pass
print(prime_numbers)
```

```python
print("Task 2.2:")
prime_indices = np.array(prime_numbers)

prime_elements = arr[prime_indices]
print("Indices used:\n", prime_indices)
print("Elements at prime indices:\n", prime_elements)
```

3. Replace values at indices divisible by 3 with -1.

```python
print("Task 2.3:")
div_by_3_indices = np.arange(0, 30, 3) 

print(f"\n{div_by_3_indices}")
arr[div_by_3_indices] = -1
print("Array after replacement:\n", arr)
```

4. Verify the changes using NumPy boolean expressions (do not use loops)

```python
print("Task 2.4:")

all_indices = np.arange(30)

mask = (all_indices % 3 == 0)

all_replaced_correctly = np.all(arr[mask] == -1)
no_unintended_changes = np.all(arr[~mask] != -1) 

print("Were all target indices set to -1? :", all_replaced_correctly)
if all_replaced_correctly:
    print("Verification successful!")

print(f"\n{all_indices}")
print(f"\n{mask}")
print(f"\n{no_unintended_changes}")
```
